```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
salary_data = pd.read_excel("salary_data_states.xlsx")
```

```python
salary_data.head()
```

```python
salary_data['JOB_TITLE_SUBGROUP'].unique()
```

```python
#Do specific sub-types of data-related jobs have higher or lower salaries than others?

avg_salary_data=salary_data.groupby(['JOB_TITLE_SUBGROUP'],as_index=False)['PAID_WAGE_PER_YEAR'].mean()
avg_salary_data.sort_values(by =['PAID_WAGE_PER_YEAR'],ascending= False )
```

```python
#Which companies have the highest salaries for those sub-types?

salary_data[['JOB_TITLE_SUBGROUP','EMPLOYER_NAME','PAID_WAGE_PER_YEAR']].sort_values(by = ['PAID_WAGE_PER_YEAR'],ascending= False ).groupby(['JOB_TITLE_SUBGROUP']).head(1)
```

```python
salary_data[['WORK_STATE','JOB_TITLE_SUBGROUP','EMPLOYER_NAME','PAID_WAGE_PER_YEAR']].sort_values(by = ['PAID_WAGE_PER_YEAR'],ascending= False ).groupby(['JOB_TITLE_SUBGROUP']).head(1)
```

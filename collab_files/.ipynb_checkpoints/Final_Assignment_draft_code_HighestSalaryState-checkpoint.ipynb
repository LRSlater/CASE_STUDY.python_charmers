```python
#What states (of those I am willing to move to) have the highest paying data-related salaries?



import pandas as pd
import numpy as np


salary_data = pd.read_csv("salary_data_states.csv")
salary_data.head()
```

```python
 salary_data.groupby('WORK_STATE').get_group('California')
```

```python
salary_data.groupby('WORK_STATE').count()
```

```python
list(salary_data['JOB_TITLE_SUBGROUP'].unique())
```

```python
#grouping data by state and job
salary_data.groupby(by=['WORK_STATE', 'JOB_TITLE_SUBGROUP']).agg(len)
```

```python
#mean salary for each state and job

sal_table = pd.pivot_table(salary_data, 
                       values='PAID_WAGE_PER_YEAR', 
                       index=['JOB_TITLE_SUBGROUP'],
                       columns=['WORK_STATE'], 
                       aggfunc=np.mean)
sal_table
```

```python
sal_table = pd.pivot_table(salary_data, 
                       values='PREVAILING_WAGE_PER_YEAR', 
                       index=['JOB_TITLE_SUBGROUP'],
                       columns=['WORK_STATE'], 
                       aggfunc=np.mean)
sal_table
```

```python
sal_table.query('JOB_TITLE_SUBGROUP == ["data analyst", "data scientist"]')




```

```python
mal_table = pd.pivot_table(salary_data, 
                       values='PREVAILING_WAGE_PER_YEAR', 
                       index=['WORK_STATE','JOB_TITLE_SUBGROUP'],
                    
                       aggfunc=np.mean)
mal_table

```

```python
#What states (of those I am willing to move to) have the highest paying(prevailing wage/year)  data-related salaries? Data Analysts


dafr = mal_table.query('JOB_TITLE_SUBGROUP == ["data analyst"]').sort_values(by=['PREVAILING_WAGE_PER_YEAR'], ascending = False)

dafr.columns

```

```python
ndf = salary_data[['PREVAILING_WAGE_PER_YEAR','WORK_STATE','JOB_TITLE_SUBGROUP']].copy()
print(ndf)
```

```python

from matplotlib import pyplot as plt
x = ndf['WORK_STATE']
y = ndf['PREVAILING_WAGE_PER_YEAR']

plt.scatter(x, y)
plt.show()
```

```python
#What states (of those I am willing to move to) have the highest paying (prevailing wage/year) data-related salaries? Data Scientists
mal_table.query('JOB_TITLE_SUBGROUP == ["data scientist"]').sort_values(by=['PREVAILING_WAGE_PER_YEAR'], ascending = False)
```

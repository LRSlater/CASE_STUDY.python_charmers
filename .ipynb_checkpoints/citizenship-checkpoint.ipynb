```python
import pandas as pd
import numpy as np


salary_data = pd.read_csv("salary_data_states.csv")
salary_data.head()
```

```python
salary_data.info()
```

```python
salary_data['COUNTRY_OF_CITIZENSHIP'].value_counts()

```

```python
#salary_data.groupby(by=[ 'CASE_STATUS', 'COUNTRY_OF_CITIZENSHIP']).agg(len)


citizen = pd.pivot_table(salary_data, 
                         index= 'CASE_STATUS', columns = 'COUNTRY_OF_CITIZENSHIP', aggfunc = 'count')

citizen
```

```python

certified_row = citizen.iloc[[0]]
certified_row
```

```python
highest_cert = (certified_row['CASE_NUMBER'] > 10)

highest_cert
```

```python
cert = pd.DataFrame(certified_row['CASE_NUMBER'] > 100)

pd.set_option("display.max_rows", None, "display.max_columns", None)
cert

# 50 and over: CANADA, CHINA, FRANCE, GERMANY, INDIA, MEXICO, NEPAL, PHILIPPINES, RUSSIA, SOUTH KOREA, TAIWAN
# 70 and over: CANADA, CHINA, INDIA, MEXICO, PHILIPPINES, SOUTH KOREA, TAIWAN
# 100 and over: CANADA, CHINA, INDIA, MEXICO, SOUTH KOREA
```
